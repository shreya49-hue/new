```python
import numpy as np

# Dataset: 10 students, 5 subjects (marks out of 100)
marks = np.array([
    [78, 85, 90, 88, 76],
    [92, 80, 75, 85, 89],
    [65, 70, 72, 68, 60],
    [88, 90, 92, 86, 84],
    [55, 60, 65, 58, 62],
    [75, 78, 80, 82, 79],
    [95, 88, 90, 94, 92],
    [60, 65, 70, 68, 66],
    [82, 85, 88, 80, 83],
    [70, 72, 74, 68, 71]
])
```

```python
avg_student = np.mean(marks, axis=1)
print(avg_student)
```

```python
avg_subject = np.mean(marks, axis=0)
print(avg_subject)
```

```python
highest_student = np.max(marks, axis=1)
print(highest_student)
```

```python
lowest_student = np.min(marks, axis=1)
print(lowest_student)
```

```python
all_sub = np.all(marks >80, axis=1)
print(all_sub)
```

```python
all_sub = np.all(marks >= 60, axis=1)
print(all_sub)
```

```python
marks_grace = marks + 5
print(marks_grace)
```

```python
reshaped = marks.reshape(5, 10)
print(reshaped)
```

```python
flattened = reshaped.flatten()
print(flattened)
```

```python
total_marks = np.sum(marks, axis=1)
print(total_marks)
```

```python
top_student_index = np.argmax(total_marks)
print(top_student_index)
```
